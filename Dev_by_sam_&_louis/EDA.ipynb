### Distribution of the target variable 

```python
# Distribution of the target variable
sns.histplot(df_train['log_bike_count'], kde=True)
plt.title('Distribution of Log Bike Count')
plt.show()
```

### Daily average log Bike Count overall

```python
# Daily average log bike count :
df_train.set_index('date', inplace=True)
df_train['log_bike_count'].resample('D').mean().plot(title='Daily Average Log Bike Count')
plt.show()
```

### Bike Count over time for each of the top 10 station

```python

# Extract the week start dates (or other desired grouping periods)
df_train.reset_index(inplace=True)
df_train['week'] = df_train['date'].dt.to_period('W').apply(lambda r: r.start_time)

# Group by 'week' and 'site_name' (station)
pivot_df = (
    df_train.groupby(['week', 'site_name'])['bike_count']
    .sum()
    .unstack()
)

# Plot the data
pivot_df.plot(figsize=(12, 6), linewidth=2)  # Ensure line graph with clear visibility
plt.title("Bike Count by Date", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Bike Count", fontsize=12)
plt.legend(title="Station", bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=10, frameon=False)
plt.tight_layout()
plt.show()
```

### Log Bike count for the top 10 sites 

```python
# Analysis of bike count by site
top_sites = df_train['site_name'].value_counts().head(10).index
top_sites_data = df_train[df_train['site_name'].isin(top_sites)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='site_name', y='log_bike_count', data=top_sites_data, order=top_sites)
plt.xticks(rotation=45)
plt.title('Log Bike Count by Site (Top 10 Sites)')
plt.show()
```

### Spatial distribution of stations with their log bike counts 

```python
# Spatial distribution of bike counters
plt.figure(figsize=(10, 8))
plt.scatter(df_train['longitude'], df_train['latitude'], c=df_train['log_bike_count'], cmap='viridis', alpha=0.5)
plt.colorbar(label='Log Bike Count')
plt.title('Spatial Distribution of Bike Counts')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### Log Bike Count depending on installation year of stations

```python
# Installation year analysis
df_train['installation_year'] = df_train['counter_installation_date'].dt.year
plt.figure(figsize=(12, 6))
sns.boxplot(x='installation_year', y='log_bike_count', data=df_train)
plt.title('Log Bike Count by Installation Year')
plt.show()
```

### Date / Time trend analysis 

```python
# For exploration purposes, we create new columns with date features :
df_train.reset_index(inplace=True)

df_train['year'] = df_train['date'].dt.year
df_train['month'] = df_train['date'].dt.month
df_train['day'] = df_train['date'].dt.day
df_train['weekday'] = df_train['date'].dt.weekday
df_train['hour'] = df_train['date'].dt.hour
```

```python
# Trends by hour of log bike count overall
sns.boxplot(x='hour', y='log_bike_count', data=df_train)
plt.title('Hourly Distribution of Log Bike Count')
plt.show()
```

```python
# Pivot the data to create a matrix of weekdays and hours
heatmap_data = df_train.pivot_table(index='hour', columns='weekday', values='log_bike_count', aggfunc='mean')

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=False)
plt.title('Average Log Bike Count by Hour and Weekday')
plt.xlabel('Weekday (0=Monday, 6=Sunday)')
plt.ylabel('Hour of Day')
plt.show()
```

```python
# Same thing in a graph format :
# Group data by 'weekday' and 'hour' to compute the average log bike count
average_hourly = df_train.groupby(['weekday', 'hour'])['log_bike_count'].mean().reset_index()

# Create a pivot table for easier plotting
pivot_data = average_hourly.pivot(index='hour', columns='weekday', values='log_bike_count')

# Plot the data
plt.figure(figsize=(12, 6))
for day in range(7):
    plt.plot(pivot_data.index, pivot_data[day], label=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'][day])

# Customize the plot
plt.title('Average Log Bike Count per Hour for Each Day of the Week', fontsize=14)
plt.xlabel('Hour of Day', fontsize=12)
plt.ylabel('Average Log Bike Count', fontsize=12)
plt.xticks(range(0, 24))  # Ensure x-axis covers all 24 hours
plt.legend(title='Day of the Week', fontsize=10)
plt.grid(True)
plt.show()
```

```python
# Trafffic trends over time : monthly trends :
# Group by the 'month' column and calculate the mean log bike count for each month
weekly_traffic = df_train.groupby('month')['log_bike_count'].mean()

plt.figure(figsize=(12, 6))
weekly_traffic.plot()
plt.title('Monthly Average Log Bike Count')
plt.ylabel('Log Bike Count')
plt.xlabel('month')
plt.show()
```

```python
df_train['day_type'] = df_train['weekday'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Plot comparison
plt.figure(figsize=(8, 6))
sns.boxplot(x='day_type', y='log_bike_count', data=df_train)
plt.title('Traffic Comparison: Weekday vs Weekend')
plt.show()
```

**Correlation relationship**

```python
print(df_train.columns)
```

```python
# Create a copy of the DataFrame to transform non-numeric columns
df_temp = df_train.copy()

# Transform non-numeric variables into numeric ones
df_temp['counter_id_encoded'] = df_temp['counter_id'].astype('category').cat.codes
df_temp['counter_technical_id_encoded'] = df_temp['counter_technical_id'].astype('category').cat.codes
df_temp['coordinates_encoded'] = df_temp['coordinates'].astype('category').cat.codes
df_temp['date_numeric'] = pd.to_datetime(df_temp['date']).astype('int64') // 10**9
df_temp['counter_installation_date_numeric'] = pd.to_datetime(df_temp['counter_installation_date']).astype('int64') // 10**9

#We only select the features that interest us
features_to_include = [
    'bike_count',
    'counter_id_encoded',
    'site_id',
    'date_numeric',
    'counter_installation_date_numeric',
    'coordinates_encoded',
    'counter_technical_id_encoded',
    'latitude',
    'longitude'
]

# Calculate correlations with 'bike_count'
correlations = df_temp[features_to_include].corr()['bike_count']

# Convert correlations to a DataFrame (for visualization)
correlation_df = correlations.to_frame().reset_index()
correlation_df.columns = ['Variable', 'Correlation']  
correlation_df = correlation_df.sort_values(by='Correlation', ascending=False)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_df.set_index('Variable').T, annot=True, cmap='coolwarm', cbar=True)
plt.title('Heatmap of correlations with bike_count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Missing values**

We choose to remove all variables with more than 60% missing values, because they contain too little usable information.\
For the remaining variables like *'Ctype2'* or *'ch'*, whose values are very similar (because they are defined by zones), we impute them using the median of the other observations, even if almost 50% of the values are missing.\
For other variables with a low rate of missing values, the median is also used as an imputation method, because it is robust to outliers and preserves the overall trends of the data.

```python
# Delete columns with more than 60% missing values
columns_to_drop = missing_percentage[missing_percentage > 60].index
external_data = external_data.drop(columns=columns_to_drop)

# Replace the others missing values by the median :
columns_to_treat = external_data.columns[external_data.isnull().sum() > 0]  # Colonnes with NaN
for col in columns_to_treat:
    external_data[col].fillna(external_data[col].median(), inplace=True)

print("Percentage of missing values after processing:")
print(external_data.isnull().sum())
```
